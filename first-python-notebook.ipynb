```python
top_supporters = support.fillna("").groupby(
    ["contributor_lastname", "contributor_firstname"]
).amount.sum().reset_index().sort_values("amount", ascending=False).head(10)
```

```python
top_supporters
```

```python
top_supporters['contributor_fullname'] 
    = top_supporters.contributor_firstname + " " + top_supporters.contributor_lastname
```

```python
alt.Chart(top_supporters).mark_bar().encode(
    x='contributor_fullname',
    y='amount'
)
```

```python
alt.Chart(top_supporters.head(5)).mark_bar().encode(
    x='amount',
    y=alt.Y('contributor_fullname', sort="-x")
)
```
